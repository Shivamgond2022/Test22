# Making user itme matrics chart

```python
#using pivot tabel
book_df = book.pivot_table(index='user_id',columns='title',values='rating')
book_df.head()
```

```python
book_df.fillna(0,inplace=True)
```

```python
book_df
```

# Calculating cosine similarity between users

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book_user = 1-pairwise_distances(book_df.values,metric="cosine")
book_user
```

```python
# storeing result in dataframe
data = pd.DataFrame(book_user)
```

```python
data.index=book["user_id"].unique()
data.columns=book["user_id"].unique() 
```

```python
data.shape
```

```python
data.iloc[0:5,0:5]
```

```python
np.fill_diagonal(book_user,0)
data.iloc[0:5,0:5]
```

```python
# Most similar user
data.idxmax(axis=1)
```
